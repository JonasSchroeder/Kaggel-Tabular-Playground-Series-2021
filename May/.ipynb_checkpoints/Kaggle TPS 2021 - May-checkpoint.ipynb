```python
# Standardize numeric features

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_trans = scaler.fit_transform(X_train)
X_submit_trans = scaler.transform(X_submit)
```

## 1. Baseline Model
Start with no feature engineering a a few standard models for a baseline performance

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score


models = [
    RandomForestClassifier(n_estimators=200, max_depth=3, random_state=0),
    LinearSVC(max_iter=1000),
    LogisticRegression(random_state=0),
]

CV = 5
cv_df = pd.DataFrame(index=range(CV * len(models)))
entries = []

for model in models:
    model_name = model.__class__.__name__
    accuracies = cross_val_score(model, X_train_trans, y_train, scoring='accuracy', cv=CV)
  
    for fold_idx, accuracy in enumerate(accuracies):
        entries.append((model_name, fold_idx, accuracy))
        
    cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'accuracy'])


import seaborn as sns
sns.boxplot(x='model_name', y='accuracy', data=cv_df)
sns.stripplot(x='model_name', y='accuracy', data=cv_df, 
              size=8, jitter=True, edgecolor="gray", linewidth=2)
```

```python
cv_df.groupby('model_name').accuracy.mean()
```

## 2. Improved RF model through RandomizedSearch and GridSearch

Using scoring = roc_auc to not just focus on precision, and class_weight parameter to deal with imbalanced data.

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV

param_dist = {
    'n_estimators': np.arange(100,1500,25),
    'max_features': ['auto', 'sqrt'],
    'max_depth': np.arange(10,1000,10),
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1,2,4],
    'bootstrap': [True, False]
}

print(param_dist)
```

### a) OVO non_weighted

```python
from sklearn.ensemble import RandomForestClassifier
rnd_clf = RandomForestClassifier(class_weight='balanced')

random_search = RandomizedSearchCV(estimator=rnd_clf, 
                                   param_distributions=param_dist,
                                   scoring='roc_auc_ovo',
                                   n_iter=100,
                                   n_jobs=-1, 
                                   cv=3, 
                                   verbose=10, 
                                   random_state=42)

random_search.fit(X_train, y_train)
```

```python
print(random_search.best_params_)
```

```python
print(random_search.best_score_)
```

```python
y_pred = random_search.predict(X_test)

from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(conf_mat, annot=True, fmt='d')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.savefig("img_export/confusion_matrix_ovo_nonweighted_absolute.jpg")
```

```python
# Plot normalized confusion matrix
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(random_search, X_test, y_test, display_labels=['Class_1','Class_2','Class_3','Class_4'], 
                      cmap=plt.cm.Blues, normalize="true")
plt.title("Normalized Confusion Matrix for OvO non_weighted Random Forest\n", size=14)
plt.savefig("img_export/confusion_matrix_ovr_nonweighted_normalized.jpg")
```

## 3. Experimenting with simple DNNs

```python
X = train_df.drop(columns=["target", "id"])
y = train_df.target

X_submit = submit_df.drop(columns=["id"])
```

```python
# Standardize numeric features

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_trans = scaler.fit_transform(X)
X_submit_trans = scaler.transform(X_submit)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X_trans, y, test_size=0.2, shuffle=True, random_state=0)
```

```python
# encode from y column to four
from sklearn.preprocessing import LabelBinarizer

enc = LabelBinarizer()
enc.fit(y)

y_train_encoded = enc.transform(y_train)

y_valid_encoded = enc.transform(y_valid)
```

```python
X_train = pd.DataFrame(X_train)
y_train_encoded = pd.DataFrame(y_train_encoded)

X_valid = pd.DataFrame(X_valid)
y_valid_encoded = pd.DataFrame(y_valid_encoded)

X_submit = pd.DataFrame(X_submit_trans)
```

```python
import tensorflow as tf
from tensorflow import keras

# Model definition
model = keras.models.Sequential([
    keras.layers.Dense(100, activation="relu", kernel_initializer='he_uniform'),
    keras.layers.Dense(100, activation="relu", kernel_initializer='he_uniform'),
    keras.layers.Dense(100, activation="relu", kernel_initializer='he_uniform'),
    keras.layers.Dense(100, activation="relu", kernel_initializer='he_uniform'),
    keras.layers.Dense(100, activation="relu", kernel_initializer='he_uniform'),
    keras.layers.Dense(100, activation="relu", kernel_initializer='he_uniform'),
    keras.layers.Dense(4, activation="softmax")
])

# specify loss function and optimizer
model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["accuracy", "AUC"])
```

```python
# fit the model
history = model.fit(X_train, y_train_encoded, epochs=100, 
                    validation_data=(X_valid, y_valid_encoded),
                    callbacks=[keras.callbacks.EarlyStopping(patience=5)])
```

```python
# plot learning curves
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0.55,0.85)
```

```python
model.evaluate(X_valid, y_valid_encoded)
```

```python
model.predict(X_submit)
```

```python
import datetime as datetime

y_submit = model.predict(X_submit)

export_df = pd.DataFrame()
export_df["id"] = pd.read_csv("input/test.csv")["id"]
export_df["Class_1"] = y_submit[:,0]
export_df["Class_2"] = y_submit[:,1]
export_df["Class_3"] = y_submit[:,2]
export_df["Class_4"] = y_submit[:,3]
now = datetime.datetime.now()
name_add = "date_"+str(now.year)+"-"+str(now.month)+"-"+str(now.day)+"_time_"+str(now.hour)+"-"+str(now.minute)
export_df.to_csv(f"output/DNN_{name_add}.csv", index=False)
```

### GridSearch Hyperparameter for DNN

```python
# Classification model
def build_model(n_hidden=2, n_neurons=100, learning_rate=0.003):
    model = keras.models.Sequential()
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation="relu", kernel_initializer='he_uniform'))
    model.add(keras.layers.Dense(4, activation="softmax"))
    optimizer = keras.optimizers.Adam(lr=learning_rate)
    model.compile(loss="categorical_crossentropy", optimizer=optimizer, metrics="AUC")
    return model
```

```python
# scikit-learn wrapper around keras model
keras_clf = keras.wrappers.scikit_learn.KerasClassifier(build_model, verbose=True)
```

```python
%%time

from sklearn.model_selection import GridSearchCV

param_distribs = {
    #"n_hidden": [0,1,2,3,4],
    #"n_neurons": np.arange(10,500,10).tolist(),
    "learning_rate": np.arange(0.0001, 0.3, 0.005).tolist(),
}

# GridSearch
grid_search_cv = GridSearchCV(keras_clf, param_distribs, scoring="roc_auc", cv=3, n_jobs=-1, verbose=1)
grid_search_cv.fit(X_train, y_train_encoded, epochs=50, verbose=1,
                  validation_data=(X_valid, y_valid_encoded),
                  callbacks=[keras.callbacks.EarlyStopping(patience=5)])
```

```python
grid_search_cv.cv_results_
```

```python
scores = pd.DataFrame()
scores["learning_rate"] = param_distribs["learning_rate"]
scores["mean_test_score"] = grid_search_cv.cv_results_["mean_test_score"]
scores["rank_test_score"] = grid_search_cv.cv_results_["rank_test_score"]
scores
```

```python
print("Best hyperparameters:")
print(grid_search_cv.best_params_)
print("\nBest score:")
print(grid_search_cv.best_score_)
```

### GridSearch and Hyperparameter tuning for LightGBM classifier

Parameters to tune:
- max_bin
- num_leaves
- num_iterations
- learning_rate
- max_depth
- bagging_fraction
- feature_fraction

```python
%%time
# max_bin

import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.metrics import f1_score

max_bin = np.arange(2,100,2)

scores = pd.DataFrame(columns=["max_bin", "train_f1", "test_f1"])

for i in range(len(max_bin)):
    scores.at[i, "max_bin"] = max_bin[i]
    lgbm_clf = LGBMClassifier(max_bin=max_bin[i])
    lgbm_clf.fit(X_train, y_train)
    scores.at[i, "train_f1"] = round(f1_score(y_train, lgbm_clf.predict(X_train), average="micro"),4)
    scores.at[i, "test_f1"] = round(f1_score(y_test, lgbm_clf.predict(X_test), average="micro"),4)

scores.head(10)
```

```python
import seaborn as sns

scores["train_f1"] = scores.train_f1.astype(float)
scores["test_f1"] = scores.test_f1.astype(float)

sns.lineplot(scores["max_bin"], scores["train_f1"], label="Training Score")
sns.lineplot(x=scores["max_bin"], y=scores["test_f1"], label="Validation Score")
plt.title("Scores depending on max_bin")
```

```python
%%time

# num_leaves
num_leaves = np.arange(2,100,1)

scores = pd.DataFrame(columns=["num_leaves", "train_f1", "test_f1"])

for i in range(len(num_leaves)):
    scores.at[i, "num_leaves"] = num_leaves[i]
    lgbm_clf = LGBMClassifier(num_leaves=num_leaves[i])
    lgbm_clf.fit(X_train, y_train)
    scores.at[i, "train_f1"] = round(f1_score(y_train, lgbm_clf.predict(X_train), average="micro"),4)
    scores.at[i, "test_f1"] = round(f1_score(y_test, lgbm_clf.predict(X_test), average="micro"),4)

scores.head(10)
```

```python
import seaborn as sns

scores["train_f1"] = scores.train_f1.astype(float)
scores["test_f1"] = scores.test_f1.astype(float)

sns.lineplot(scores["num_leaves"], scores["train_f1"], label="Training Score")
sns.lineplot(x=scores["num_leaves"], y=scores["test_f1"], label="Validation Score")
plt.title("Scores depending on num_leaves")
```
